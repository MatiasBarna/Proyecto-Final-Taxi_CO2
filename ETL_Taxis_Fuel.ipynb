# Diccionario de Datos

- **Year:** Año de fabricación del vehículo.
- **Manufacturer:** Marca del vehículo.
- **Model:** Modelo del vehículo.
- **CO2 (p/mile):** Emisiones de CO2 por milla.
- **Miles per gallon (mpg):** Consumo de combustible en millas por galón.
- **ef_x_comb:** Puntuación de eficiencia del vehículo por combustible.
- **comb_x_year:** Gasto estimado de combustible por año.
- **Fuel:** Tipo de combustible utilizado por el vehículo.
- **score_ghg:** Puntuación de emisiones de gases de efecto invernadero (GHG) por vehículo (Cuanto mayor sea la puntuación, menos emisiones de gases de efecto invernadero emitirá el vehículo.).
- **ef_x_comb_city:** Eficiencia de combustible en condiciones de conducción en la ciudad.
- **youSaveSpend:** Indica cuánto se puede ahorrar o gastar en combustible en comparación con un vehículo promedio.
- **Alternative Fuel:** Tipo de combustible alternativo que puede utilizar.
- **createdOn:** Indica la fecha de creación de los registros en el conjunto de datos.
- **modifiedOn:** Indica la fecha de modificación de los registros en el conjunto de datos.

```python
# Se cuentan cuantos nulos hay por columna
df_taxis_fuel.isnull().sum()
```

```python
# Se eliminan las filas donde 'Fuel' es nulo
df_taxis_fuel = df_taxis_fuel[df_taxis_fuel['Fuel'].notna()]
df_taxis_fuel.head(2)
```

```python
# Completamos los valores nulos de la columna 'Alternative Fuel' con 'No', ya que se trata de los vehiculos no hibridos.
df_taxis_fuel['Alternative Fuel'] = df_taxis_fuel['Alternative Fuel'].fillna('No')
df_taxis_fuel.head(2)
```

```python
# Se revisan duplicados
duplicados= df_taxis_fuel.duplicated().sum()
print(f'cantidad de datos duplicados {duplicados}')
```

```python
# Se eliminan las filas con duplicados
df_taxis_fuel = df_taxis_fuel.drop_duplicates()
duplicados= df_taxis_fuel.duplicated().sum()
print(f'cantidad de datos duplicados {duplicados}')
```

```python
# Se cuentan la cantidad de datos con valor -1 en la columna 'CO2 (p/mile)'
conteo_valores_menos_uno = df_taxis_fuel['CO2 (p/mile)'].eq(-1).sum()

print(f"El número de valores -1 en la columna 'CO2 (p/mile)' es: {conteo_valores_menos_uno}")
```

```python
# Se filtra el DataFrame para ver solo las filas donde la columna 'CO2 (p/mile)' es -1
df_filtrado = df_taxis_fuel[df_taxis_fuel['CO2 (p/mile)'] == -1]

# Se muestra el DataFrame filtrado
df_filtrado
```

```python
# Se filtra el DataFrame para ver solo las filas donde la columna 'Fuel' es 'Electricity'
df_filtrado = df_taxis_fuel[df_taxis_fuel['Fuel'] == 'Electricity']

# Se muestra el DataFrame filtrado
df_filtrado
```

```python
# Se calcula el promedio de la columna "co2" sin contar los valores -1
promedio_co2 = df_taxis_fuel[df_taxis_fuel['CO2 (p/mile)'] != -1]['CO2 (p/mile)'].mean()
promedio_co2 = round(promedio_co2,2)
promedio_co2
```

```python
# Se reemplaza los valores -1 por el promedio
df_taxis_fuel['CO2 (p/mile)'] = df_taxis_fuel['CO2 (p/mile)'].replace(-1, promedio_co2)
df_taxis_fuel
```
